```python
import nltk
from wordcloud import WordCloud, STOPWORDS
from PIL import Image # 그림을 불러오는 패키지
import numpy as np
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
import pickle
```

```python
with open('../data1/A_new_hope.txt') as file:
    text = file.read()
```

```python
sw_wc = STOPWORDS
```

```python
sw_wc.add('said')
sw_wc.add('int')
sw_wc.add('ext')
```

```python
wc = WordCloud(background_color='white', max_words=2000,
               stopwords=sw_wc)
wc = wc.generate(text)
```

```python
count = 0
for key, value in wc.words_.items():
    print(key, ':', value)
    count += 1
    if count > 10:
        break
```

```python
plt.figure(figsize=(10,6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
