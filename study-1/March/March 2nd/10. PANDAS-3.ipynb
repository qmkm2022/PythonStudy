```python
df_ns = df.iloc[[0, 5], 3:] #1990년도 부터 남북 합계 행 2개 가져오기
df_ns.index = ['south', 'north'] #행 인덱스 변경
df_ns.columns = df_ns.columns.map(int) #열 이름을 정수형으로 변경
df_ns.head()
```

```python
# 선그래프 그리기
df_ns.plot()
```

```python
# 행과 열을 전치하여 다시 그리기
tdf_ns = df_ns.T
tdf_ns.plot()
```
