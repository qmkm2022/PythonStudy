```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import folium
import re
import json
from matplotlib import font_manager, rc
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family = 'Malgun Gothic')
```

```python
df = pd.read_csv('./data1/서울시 주요 공원현황.csv', encoding='EUC-KR')
df.head()
```

# 1. 공원 면적 정리

```python
area = df.면적.astype(str).apply(lambda x : x.strip('총 ').split()[0])
area[40:30]
```

```python
area.unique()
```

```python
area[(area == 'nan') | (area == '휴양')] = '0㎡'
```

```python
area.unique() # 단위표시가 모두 다르기 때문에 정교표현식으로 제거
```

```python
area = area.apply(lambda x : float(re.sub('[m² ㎡ ㎥]', '', x)))
```

```python
area = area.apply(lambda x : int(np.round(x)))
```

# 2. 공원의 면적에 따라 분류

```python
area_criteria = [-1, 100000, 1000000, 12000000 ]
labels = ['소형', '중형', '대형']
size_info = [3, 7, 15]
scale = pd.cut(area, area_criteria, labels=labels)
size = pd.cut(area, area_criteria, labels=size_info)
```

```python
size
```

```python
scale
```

# 3. 새로운 프레임 만들기

```python
df = df[['공원명', '지역', 'X좌표(WGS84)', 'Y좌표(WGS84)']]
df.columns=['공원명', '지역', '경도', '위도']
```

```python
df.head(3)
```

```python
df['면적'] = area
df['분류'] = scale
df['크기'] = size
df.head()
```

```python
df.isnull().sum()
```

```python
df[df.지역.isna()]
```

```python
df[df.경도.isna()]
```

```python
df = df.dropna()
```

```python
df.to_csv('서울공원요약.csv', index=False)
```

# 4. 공원 시각화

```python
map = folium.Map(location=[df.위도.mean(), df.경도.mean()], zoom_start=12)
for i in df.index:
    folium.Circle( radius=int(df.크기[i]), location=[df.위도[i], df.경도[i]], popup=folium.Popup(df.면적[i], max_width=200),
                  tooltip=f"{df.공원명[i]}({df.면적[i]: ,d})㎡",
                  color = 'crimson', fill=True, fill_color = 'crimson').add_to(map)
title = '<h3 align="center" style="font-size:20"> 서울시 공원</h3>'
map.get_root().html.add_child(folium.Element(title))
map
```
