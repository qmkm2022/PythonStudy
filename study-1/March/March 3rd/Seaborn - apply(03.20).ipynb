```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.info
```

```python
# 1. age를 평균으로 대체
# 2. embarked 최빈값으로 대체
# 3. embak_ town 결측값이 있는 행 삭제
# 4. deck 삭제
#5. titanic['adult/child'] 컬럼을 만들고 20살 이상이면 adiut 이하면 child apply와
#   lambda를 활용
# 6. bins = [1, 20, 30, 50, 70, 100]
# labels = ['미성년자', '청년', '중년', '장년', '노년]
# age를 기준으로 'age_cut'컬럼을 생성 후 두 컬럼을 출력하기
# age, age_cut
# 22 청년
# 38 중년
# pd.cut
```

```python
# 1. age를 평균으로 대체
titanic.age.fillna(titanic.age.mean(), inplace=True)
```

```python
# 2. embarked 최빈값으로 대체
titanic.embarked.value_counts()
```

```python
titanic.embarked.fillna('S', inplace=True)
```

```python
# 3. embak_ town 결측값이 있는 행 삭제
titanic.dropna(subset=['embark_town'], inplace=True)
```

```python
# 4. deck 삭제
titanic.drop(columns=['deck'], inplace=True)
```

```python
titanic.shape
```

```python
#5. titanic['adult/child'] 컬럼을 만들고 20살 이상이면 adiut 이하면 child apply와
#   lambda를 활용
titanic['adult/child'] = titanic.age.apply(lambda x: 'adult'
                                          if x > 20 else 'child')
```

```python
titanic.tail()
```

```python
# 6. bins = [1, 20, 30, 50, 70, 100]
# labels = ['미성년자', '청년', '중년', '장년', '노년']
# age를 기준으로 'age_cut'컬럼을 생성 후 두 컬럼을 출력하기
# age, age_cut
# 22 청년
# 38 중년
# pd.cut
bins = [1, 20, 30, 50, 70, 100]
labels = ['미성년자', '청년', '중년', '장년', '노년']
titanic['age_cut'] = pd.cut(titanic.age, bins, labels=labels)
```

```python
df = pd.read_csv('./data1/scores.csv')
df = df.melt(id_vars = 'name', var_name = 'subject', value_name='score')
df
```

```python
def get_grade(x):
    if x > 90 :
        grade = 'A'
    elif x >= 80:
        grade = 'B'
    elif x >= 70:
        grade = 'C'
    elif x >= 60:
        grade = 'D'
    else:
        grade = 'F'
    return grade
```

```python
df['grade'] = df['score'].apply(get_grade)
df = df.sort_values('name')
df
```

```python
# 데이터 프레임.pivot(index = 인덱스로 사용할 컬럼,
# columns = 컬럼으로 사용할 컬럼, values=값으로 사용할 컬럼)
# name, subject, score
df.pivot(index='name', columns='subject', values='grade')
```

# Groupby

```python
# df.groupby(그룹기준컬럼).통계적용컬럼.통계함수
# .count() : 누락값을 제외한 데이터수
# .size() : 누락값을 포함한 데이터수
# .mean(), .sum(), .min(), .max()
```

```python
# 객실등급(pclass) 별 승선자수 결과 집계 -count함수
df_titanic1 = df_titanic.groupby('pclass').survived.count().to_frame()
df_titanic1
```

```python
# 객실등급(pclass) 별 생존자수 결과 집계 - sum함수
df_titanic2 = df_titanic.groupby('pclass').survived.sum().to_frame()
df_titanic2
```

```python
#객실등급(pclass) 별 생존비율 결과 집계
df_titanic3 = df_titanic.groupby('pclass').survived.mean().to_frame()
df_titanic3
```

```python
df_titanic4 = pd.concat([df_titanic1, df_titanic2, df_titanic3], 
                      axis=1)
df_titanic4.columns = ['승선자수', '생존자수', '생존비율']
df_titanic4
```

```python
df_titanic = sns.load_dataset('titanic')
df_titanic = df_titanic[['survived','pclass','sex','age','embarked']]
df = df_titanic.dropna()
df.head()
```

```python
# 성별 생존통계
df5 = df.groupby('sex').survived.count().to_frame()
df5
```

```python
# 성별 생존통계
df6 = df.groupby('sex').survived.sum().to_frame()
df6
```

```python
# 성별 생존율
df7 = df.groupby('sex').survived.mean().to_frame()
df7
```

```python
# 성별 탑승자수, 생존자수, 생존율
df8 = pd.concat([df5,df6,df7], axis = 1)
df8.columns = ['승선자수', '생존자수', '생존비율']
df8
```

```python
# 성별, 객실등급별 생존율
df.groupby(['sex', 'pclass']).survived.mean().to_frame
```

```python
# 내가만든(사용자정의)함수 적용 .agg(사용자 정의 함수, 매개변수)
# 사용자 함수 선언
def my_mean(values):
    return sum(values) / len(values)
```

```python
df.groupby(['sex','pclass']).survived.agg(my_mean)
```
