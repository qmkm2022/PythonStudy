```python
import pandas as pd

f = open("../data/text/news1.txt", encoding='UTF-8')
lines = f.readlines()
f.close()
```

```python
import re
from konlpy.tag import Hannanum

han = Hannanum()
word_list = []
for i in range(len(lines)):
    sentence =re.sub('[^가-힣ㄱ-ㅎㅏ-ㅣa-zA-Z]', ' ', lines[i])
    # 명사만 추출
    a=sentence.strip()
    n=han.nouns(a)
    n2=[x for x in n if len(x)>1]
    if len(n2)>0:
        word_list.append(n2)

# tf-idf 행렬을 만들기 위해 다시 역토큰화
detokenized_doc = []
for i in range(len(word_list)):
    t = ' '.join(word_list[i])
    detokenized_doc.append(t)

detokenized_doc
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(detokenized_doc)
X.shape # TF-IDF 행렬의 크기 확인
```

```python
from sklearn.decomposition import TruncatedSVD
# n_components 토픽수
svd_model = TruncatedSVD(n_components=5)
svd_model.fit(X)
len(svd_model.components_)
```

```python
import numpy as np
# 토픽수 x 단어수
np.shape(svd_model.components_)
```

```python
svd_model.components_
```

```python
# 단어 집합
terms = vectorizer.get_feature_names_out()

def get_topics(components, feature_names, n=5):
    for idx, topic in enumerate(components):
        print("Topic %d:" % (idx+1),
              [(feature_names[i], topic[i].round(5)) for i in topic.argsort()[:-n - 1:-1]])

get_topics(svd_model.components_,terms)
# 각 토픽의 핵심 키워드 추출
```
