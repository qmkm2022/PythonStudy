```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ls1 = [3400, 2300, 1200, 2110, 3320, 1220, 2000, 4600, 2300, 3400]

ls2 = [2100, 2200, 1800, 2220, 3110, 1420, 2100, 4530, 2200, 3100]

ls3 = [3300, 3000, 2100, 2240, 3530, 1620, 2200, 4200, 1300, 4200]

ls4 = [3300, 2700, 900, 2140, 3200, 2020, 2800, 4000, 4300, 4000]
```

```python
dates = pd.date_range('2020-01-01', '2020-01-10')
ls_1 = pd.Series(ls1, index = dates)
ls_2 = pd.Series(ls2, index = dates)
ls_3 = pd.Series(ls3, index = dates)
ls_4 = pd.Series(ls4, index = dates)
```

```python
df = pd.DataFrame(
    {
        'Seoul' : ls_1,
        'Busan' : ls_2,
        'Daegu' : ls_3,
        'Kwangju' : ls_4
    }
)
df
```

```python
df['Total_sell'] = df['Seoul'] + df['Busan'] + df['Daegu'] + df['Kwangju']
df
```

```python
df['Seoul - Busan'] = df['Seoul'] - df['Busan']
df
```

```python
df.plot(y='Total_sell')
```
